```python
x = [463,293]
lab = ['field','bat']
```

```python
plt.figure(figsize=(10,6))
plt.pie(x,labels=lab,
        startangle=90,
        counterclock=True,
        shadow=True,
        autopct='%1.2f%%',
       colors=['orange','cyan'],explode=(0,0.1))
plt.legend()
plt.title('Toss decision')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.hist(matches['season'],rwidth=0.9)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='season',data=matches)
plt.title('Seasons Count')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='winner',data=matches)
plt.title('Most wins by a team')
plt.xticks(rotation='vertical')
plt.xlabel('Team Name')
plt.ylabel('Win count')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='venue',data=matches)
plt.title('Matches at the stadium')
plt.xticks(rotation='vertical')
plt.xlabel('Stadium Name')
plt.ylabel('Matches Count')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='season',data=matches,hue='toss_decision')
plt.title('Toss decision varied according to season')
plt.xticks(rotation='vertical')
plt.xlabel('Season')
plt.show()
```

```python
plt.figure(figsize=(12,6))
axes = matches['player_of_match'].value_counts().head(10).plot.bar(width=0.9,color=sns.color_palette('inferno'))

for patch in axes.patches:
    axes.annotate(format(patch.get_height()),(patch.get_x() + 0.3, patch.get_height() + 0.3))
```
